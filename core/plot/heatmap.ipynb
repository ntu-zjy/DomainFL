```python
import seaborn as sns
import matplotlib.pyplot as plt
import json
import pandas as pd
```

```python
def plot_heatmap(method):
    file = f'../results/{method}/ViT-B-32_data2.json'
    results = []

    with open(file, 'r') as file:
        for line in file:

            if line.strip():
                json_data = json.loads(line)
                results.append(json_data)

    print(results[-1]['acc'])

    # draw the heatmap
    df = pd.DataFrame(results[-1]['acc'], columns = [f'Data {i} ' for i in range(1, len(results[-1]['acc'][0]) + 1)], index=[f'Client {i}' for i in range(1, len(results[-1]['acc']) + 1)])
    sns.set_context("paper")
    plt.figure(dpi=200)
    sns.heatmap(df, annot=True, fmt=".2f", cmap='Blues', vmin=0, vmax=100)
    plt.yticks(rotation=0)
    plt.xlabel('Dataset')
    plt.ylabel('Client')
    plt.title(f'Accuracy ({method})')
    plt.show()
```

```python
plot_heatmap('fedavg')
```

```python
plot_heatmap('local')
```

```python
plot_heatmap('fedts')
```
